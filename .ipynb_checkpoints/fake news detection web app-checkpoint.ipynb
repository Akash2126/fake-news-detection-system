```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (or load your CSV file)
data = {'Headline': ['News 1', 'News 2', 'News 3', 'News 4'],
        'Label': ['FAKE', 'REAL', 'FAKE', 'REAL']}
df = pd.DataFrame(data)

# Create the figure
fig, ax = plt.subplots(figsize=(6, 4))

# Bar chart
label_counts = df['Label'].value_counts()
label_counts.plot(kind='bar', ax=ax, color=['orange', 'blue'])
ax.set_title('News Label Distribution')
ax.set_ylabel('Count')

# Create a table
table_data = df.values
column_labels = df.columns

# Adding a table below the plot
table = plt.table(cellText=table_data,
                  colLabels=column_labels,
                  cellLoc='center',
                  loc='bottom',
                  bbox=[0, -0.3, 1, 0.2])  # Adjust bbox to control table position

# Style the table
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Show the plot and the table
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('news.csv')

# Display the first few rows
df.head()
```

```python
# Count of fake vs real news
label_counts = df['label'].value_counts()

# Bar chart
plt.figure(figsize=(6, 4))
label_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribution of Fake vs Real News')
plt.xlabel('News Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Boxplot for word count by label
plt.figure(figsize=(8, 5))
sns.boxplot(x='label', y='word_count', data=df, palette="Set2")
plt.title('Word Count Distribution by News Label')
plt.xlabel('News Label')
plt.ylabel('Word Count')
plt.show()
```

#### 

```python
# Pie chart for fake vs real news distribution
plt.figure(figsize=(6, 6))
plt.pie(label_counts, labels=['FAKE', 'REAL'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Fake vs Real News Distribution')
plt.show()
```

```python
# Pairplot for numeric columns (if you have other numeric features)
sns.pairplot(df, hue='label')
plt.show()
```

```python
# Heatmap to show missing values (if there are any)
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Data Heatmap')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all text data into a single string
text_data = ' '.join(df['text'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of News Articles')
plt.show()
```

```python
labels = df.label
```

```python
labels.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["text"],labels, test_size = 0.2, random_state = 20)
```

```python
X_train.head()
```

```python
# Create a new column for word count
df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))

# Histogram of word count
plt.figure(figsize=(8, 5))
df['word_count'].plot(kind='hist', bins=50, color='purple')
plt.title('Distribution of Word Count in News Articles')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
#inisilize TfFvectorizer
vector = TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
#fit and transform
tf_train = vector.fit_transform(X_train)
tf_test = vector.transform(X_test)
```

```python
# initilise PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tf_train , y_train)
```

```python
# prediction  the test dataset
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred =pac.predict(tf_test)
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy : {round(score*100,2)}%")
```

```python
# confusion metrics
confusion_matrix(y_test, y_pred, labels=['FAKE', 'REAL'])
```

```python
# save model
import pickle
filename = 'finalized_model.pkl'
pickle.dump(pac, open(filename, 'wb'))
```
